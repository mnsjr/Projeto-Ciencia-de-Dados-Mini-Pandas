# Mini Projeto de Análise de Dados

Vamos fazer um exercício completo de pandas para um miniprojeto de análise de dados.

### O que temos?

Temos os dados de 2019 de uma empresa de prestação de serviços. 

- CadastroFuncionarios
- CadastroClientes
- BaseServiçosPrestados

Obs1: Para ler arquivos csv, temos o read_csv<br>
Obs2: Para ler arquivos xlsx (arquivos em excel normais, que não são padrão csv), temos o read_excel

### O que queremos saber/fazer?

1. Qual foi o gasto total com Folha Salarial pela empresa? <br>   
    
2. Qual foi o faturamento da empresa?<br>
       
3. Qual o % de funcionários que já fechou algum contrato?<br>
        
4. Calcule o total de contratos que cada área da empresa já fechou

5. Calcule o total de funcionários por área

6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?<br>

Obs: Lembrando que as opções mais usuais de encoding:<br>
encoding='latin1', encoding='ISO-8859-1', encoding='utf-8' ou então encoding='cp1252'

```python
import pandas as pd

clientes_df = pd.read_csv(r'CadastroClientes.csv', sep = ';')
funcionarios_df = pd.read_csv(r'CadastroFuncionarios.csv', sep = ';', decimal = ',')
servicos_df = pd.read_excel(r'BaseServiçosPrestados.xlsx')
display(clientes_df)
display(funcionarios_df)
display(servicos_df)
```

### Agora excluindo informações desnecessarias dos dataframes

```python
# retirar colunas 'Estado Civil' da tabela de funcionários
# axis = 1 faz referencia à coluna e axis = 0 faz referencia às linhas a serem apagadas

funcionarios_df = funcionarios_df.drop(['Estado Civil'], axis = 1)

display(clientes_df)
display(funcionarios_df)
display(servicos_df)
```

### 1 - Qual foi o gasto total com Folha Salarial pela empresa?
Calcular o salário total de cada funcionário, salário + benefícios + impostos, depois somamos todos os salários

```python
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
display(funcionarios_df)

print('A folha de pagamento total é de R${:,.2f}.' .format(funcionarios_df['Salario Total'].sum()))
```

### 2 - Qual foi o faturamento da empresa?
Calculamos o faturamento total de cada serviço e depois somamos o faturamento de todos

Tabela Fato: é a tabela onde ficam armasenadas as informações de rotina, onde alguns itens, ex, id cliente, podem se repetir

Tabela Caracteristica: aqui ficam as caracteristicas de um produto ou perfil de cliente, tabelas com info únicos

Mesclando tabela: sempre adicionamos a Tabela Caracteristica à Tabela Fato, ainda podemos selecionar somente as colunas relevantes

```python
calcula_faturamento = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente', 'Valor Contrato Mensal']], on = 'ID Cliente')
display(calcula_faturamento)

# agora podemos criar uma nova coluna multiplicando o valor mensal pelo tempo de contrato
calcula_faturamento['Total'] = calcula_faturamento['Tempo Total de Contrato (Meses)'] * calcula_faturamento['Valor Contrato Mensal']
display(calcula_faturamento)
print('O faturamento total da empresa foi de R${:,.2f}.' .format(calcula_faturamento['Total'].sum()))
```

### 3 - Qual o % de funcionários que já fechou algum contrato?
Na base de serviços temos o funcionário que fechou cada serviço. Mas nem todos os funcionários que a empresa tem já fecharam algum serviço.

. Na base de funcionários temos uma lista com todos os funcionários

. Queremos calcular Qtde_Funcionarios_Fecharam_Serviço / Qtde_Funcionários_Totais

. Para calcular a qtde de funcionários que fecharam algum serviço, usamos a base de serviços e contamos quantos funcionários há ali. Mas lembrando, cada funcionário só pode ser contado uma única vez.

Podemos aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.

Ex: unicos_colunaA = dataframe['colunaA'].unique() dará como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA


```python
funcionarios_contrato_fechado = len(servicos_df['ID Funcionário'].unique()) # .unique() para excluir valores repetidos
funcionarios_total = len(funcionarios_df["ID Funcionário"])

print(f'O percentual de funcionários que já fechou contrato é de {funcionarios_contrato_fechado / funcionarios_total:.2%}.')
```

### 4 - Calcule o total de contratos que cada área da empresa já fechou

```python
contratos_por_area = servicos_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário', 'Cargo', 'Area']], on = 'ID Funcionário')
display(contratos_por_area)

# .value_counts() conta exatamento quantos itens há de cada um na coluna solicitada
contratos_por_area_qtde = contratos_por_area['Area'].value_counts()
print(contratos_por_area_qtde)
contratos_por_area_qtde.plot(kind = 'barh')
```

### 5 - Calcule o total de funcionários por área

```python
display(funcionarios_df)

qtde_funcionarios_area = funcionarios_df['Area'].value_counts()
print(qtde_funcionarios_area)
qtde_funcionarios_area.plot(kind = 'barh')
```

### 6 - Qual o ticket médio mensal (faturamento médio mensal) dos contratos?
.mean() calcula a média -> exemplo: media_colunaA = dataframe['colunaA'].mean()

```python
display(clientes_df)

faturamento_mensal = clientes_df['Valor Contrato Mensal'].mean()
print(f'O faturamento médio mensal dos contratos é de R${faturamento_mensal:,.2f}.')
```
